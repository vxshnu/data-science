```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=pd.read_csv('mpg_sns (1).csv')
df
```

```python
#KDE 
sns.histplot(df['mpg'],kde=True,color='skyblue')
plt.show()
```

```python
#box plot
sns.boxplot(x='mpg',y='origin',data=df)
```

```python
#pair plot
shows=['cylinders','mpg','horsepower','displacement']
sns.pairplot(df[shows])
plt.show()
```

```python
#scatter plot with regression line
sns.regplot(x='horsepower',y='mpg',data=df,line_kws={'color':'red'})
```

```python
#Correlation 
corr_matrix=df.corr(numeric_only=True)
sns.heatmap(corr_matrix,annot=True)
```

```python
#Violin plot
sns.violinplot(x='origin',y='mpg',data=df)
```

```python
#hex plot
sns.jointplot(x='weight',y='mpg',data=df,kind='hex',)
```
