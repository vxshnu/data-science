```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('mpg_sns (1).csv')
df
```

```python
df2=df.groupby('model_year')['mpg'].mean().reset_index()
df2
```

##### Line Plot

```python
plt.plot(df2['model_year'],df2['mpg'])
plt.xlabel('Model Year')
plt.ylabel('MPG')
plt.title('Model Year vs MPG')
plt.show()
```

##### Scatter Plot

```python
plt.scatter(df['mpg'],df['horsepower'])
plt.xlabel('MPG')
plt.ylabel('Horsepower')
plt.show()
```

##### Bar Plot

```python
sumdf=df['cylinders'].value_counts().reset_index()
plt.bar(sumdf['cylinders'],sumdf['count'])
plt.xlabel('Cylinder')
plt.ylabel('Count')
plt.show()
```

##### Histogram

```python
plt.hist(df['mpg'],edgecolor='black')
plt.show()
```

##### Box Plot

```python
sns.boxplot(x='origin',y='mpg',data=df)
plt.show()
```

##### Violin Plot

```python
plt.violinplot(dataset=[df[df['origin']==i]['mpg'] for i in df['origin'].unique()],positions=range(1,len(df['origin'].unique())+1),showmedians=True,vert=False,showextrema=True,showmeans=True)
plt.show()
```
